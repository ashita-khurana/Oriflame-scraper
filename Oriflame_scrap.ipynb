```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
driver = webdriver.Chrome()
import pandas as pd
search=pd.read_excel('Oriflame_Urls.xlsx',sheet_name='Sheet2')['Links']

#Scrap for half of the urls
search=search.head(1)

scrap_from=[]
for url in search:
    driver.get(url)
    try:
        num=driver.find_element_by_css_selector("div .js-product-listing-progress-bar")
        k=url
        while int(num.get_attribute('data-max'))!= int(num.get_attribute('data-value')):
            load_more=driver.find_element_by_css_selector("div .w-controls a")
            k=load_more.get_attribute('href')
            driver.get(k)
            num=driver.find_element_by_css_selector("div .js-product-listing-progress-bar")
        scrap_from.append(k)
    except:
        scrap_from.append(url)
    print(scrap_from[-1])
    
#Product list Scrap
def firstlevel(url):
    driver.get(url)
    for i in driver.find_elements_by_css_selector('div.w-product-box-brand'):
        Brands.append(i.text)
    for i in driver.find_elements_by_css_selector('div.w-product-box-name > span'):
        Name.append(i.text)
    for i in driver.find_elements_by_css_selector('span.product-box-price'):
        Price.append(i.text.replace('Rs.',''))
    for i in driver.find_elements_by_css_selector('div.w-product-box-price'):
        try:
            if i.find_element_by_css_selector('span.product-box-old-price'):
                MRP.append(i.find_element_by_css_selector('span.product-box-old-price').text.replace('Rs.',''))
        except:
            MRP.append('-')
    for i in driver.find_elements_by_css_selector('section .ui-product-list> a'):
        URLS.append(i.get_attribute('href'))
    for i in driver.find_elements_by_css_selector('span.average-rating'):
        Rating.append(i.find_elements_by_tag_name('span')[0].text.replace('reviews',''))
        
#URLS for each product and relevant details are pulled
product=[]
product_list=[]
Brands=[]
Name=[]
Price=[]
MRP=[]
URLS=[]
Rating=[]
for url in scrap_from:
    firstlevel(url)
product.append(list(zip(Brands,Name,Price,MRP,URLS,Rating)))
for i in product:
    for j in i:
        product_list.append(j)
        

#Product Page Scrap
def getdetails(u):
    try:
        driver.get(u)
        try:
            quantity.append(driver.find_element_by_css_selector('div.shades-section > p.shades-section__shade-capcity').text)
        except:
            quantity.append('-')
        try:
            productID.append(driver.find_element_by_css_selector('div.shades-section > p.shades-section__shade-code').text)
        except:
            productID.append('-')
        try:
            cat=WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, "div.breadcrumbs__wrapper"))).text.split('\n')
            if len(cat)>=2:
                Category.append(cat[1])
            else:
                Category.append("-")
            if len(cat)>=3:
                Sub_Category.append(cat[2])
            else:
                Sub_Category.append("-")
            if len(cat)>=4:
                Sub_Category2.append(cat[3])
            else:
                Sub_Category2.append("-")
        except:
            Category.append("-")
            Sub_Category.append("-")
            Sub_Category2.append("-")
        try:
            reviews.append(driver.find_element_by_css_selector('div.rating-stars > a').text.split(" ")[2])
        except:
            reviews.append('-')
    except:
        quantity.append('-')
        productID.append('-')
        Category.append("-")
        Sub_Category.append("-")
        Sub_Category2.append("-")
        
        
#Relevant product specs are pulled here
quantity=[]
Category=[]
Sub_Category=[]
Sub_Category2=[]
reviews=[]
product_details=[]
productID=[]
t=0 

print(len(product_list))  #No. of product pages to be scraped

for k in product_list:
    getdetails(k[4])
    print(t)
    t=t+1

product_details.append(list(zip(productID,quantity,reviews,Category,Sub_Category,Sub_Category2)))
product_specs=[]
for i in product_details:
    for j in i:
        product_specs.append(j)

results=[]
for a,b in zip(product_list,product_specs):
    results.append(a+b)
df1= pd.DataFrame(results, columns =['Brands','Name','Price','MRP','URLS','Rating','ProductID','Quantity','Reviews','Category','SubCategory','SubCategory2'])
print(df1)


```
